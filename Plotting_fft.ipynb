# Plotting DFT

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# N Point DFT
N = 20
```

```python
n = np.arange(0, N, 1, int)
print(n)
```

## Create the Signal

```python
x = 10 * np.sin( (2*np.pi/N) * 3 * n) + 2
print(x)
plt.plot(x)
```

```python
X = np.fft.fft(x)

# This below code block removes the noise due to calculation rounding errors
X_filtered = abs(X) > 0.1
X_new = np.zeros_like(X)
for i,is_accepted_value in enumerate(X_filtered):
    if is_accepted_value:
        X_new[i] = X[i]
X = X_new


print(abs(X))
```

## Magnitude Plot

```python
plt.figure(figsize=(15,10))
plt.xticks(n)
plt.stem(n,np.abs(X))
plt.show()
```

## Phase Plot

```python
print(np.angle(X, deg=True))
plt.ylim(-180, 180)
plt.stem(n, np.angle(X, deg=True) )
```
